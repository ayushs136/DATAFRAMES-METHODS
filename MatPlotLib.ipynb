```python
%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x**2
```

```python
y
```

```python
x
```

```python
# FUNTIONAL
plt.plot(x, y, 'g-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show() # dont have to use this unless you are in Jupyter notebook
```

```python
plt.subplot(1, 2, 1)
# no. of rows and columns in args and no. of fig
plt.plot(x, y, 'g')
plt.subplot(1, 2, 2)
plt.plot( y, x, 'b')
```

```python
# object-oriented approach
```

```python
fig = plt.figure() # figure object is created
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
# it is little more involved but it worth it

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.50, 0.4, 0.3])
axes1.plot(x, y)
axes2.plot(y, x, 'r')
axes2.set_title('Smaller Plot')
axes1.set_title('Bigger Plot')
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# axes.plot(x, y)
for current_ax in axes:
    current_ax.plot(x, y)
plt.tight_layout()
    
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)# tuple unpacking here
# axes.plot(x, y)
axes[0].plot(x, y)
axes[0].set_title('First Plot')
axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

# Figure size and DOTS PER INCH(dpi)

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))

axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig.savefig('picture.png', dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, label='x vs y')
ax.plot(y, x, label='y vs x')
ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.legend(loc=(0.8, 0.8))
```

# plot appearance

```python
fig= plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='#00f0ff', linewidth=5, alpha=.5)# RGB hex code
```

```python

fig= plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#00f0ff', lw=5, alpha=.5, linestyle='solid', marker='o', markersize=10, markerfacecolor='red', markeredgewidth=3, markeredgecolor='green')
```

```python

fig= plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#00f0ff', lw=5 , ls='--')
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
