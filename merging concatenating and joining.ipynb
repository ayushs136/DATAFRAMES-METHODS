 # OUTER JOIN 

```python
week1  = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
food = pd.read_csv('Restaurant - Foods.csv')
# OUTER  joins merges two df unmatched rows from one or both tables
```

```python
week1.head(3)
```

```python
week2.head(3)
```

```python
h = (week1.merge(week2, how='outer', on='Customer ID', suffixes=['-week1', '- week2'], indicator=True))
```

```python
h._merge.value_counts()
```

```python
mask = h['_merge'].isin(['left_only', 'right_only'])
```

```python
h[mask]
```
