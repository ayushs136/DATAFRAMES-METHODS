# A Review of the .apply() Method on Single Columns

```python
bond = pd.read_csv('jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(4)
```

```python
def convert_to_string_and_add_million(number):
    return str(number) + " MILLIONS: "
```

```python
columns = ['Box Office', 'Budget', 'Bond Actor Salary']
```

```python
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_million)
```

```python
bond.head()
```

#  The .apply() Method with Row Values

```python
bond = pd.read_csv('jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(4)
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    
    if actor=='Sean Connery':
        return 'The Best'
    elif actor =='Roger Moore' and budget >40:
        return 'Enjoyable'
    else:
        return "DOnt' know"
```

```python
bond.apply(good_movie, axis=1)
```
